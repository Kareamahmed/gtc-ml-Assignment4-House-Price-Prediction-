# 2: Identify All Missing values and Handel it

```python
df.duplicated().sum()
# no duplicate values
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
df['total_bedrooms'].fillna(df['total_bedrooms'].median(),inplace =True)
```

```python
df.isnull().sum().sum()
```

# Split Data

```python
# median income => One of the most important factors affecting home prices

df['incom_cat'] =pd.cut(df['median_income'],bins = [0.,1.5,3.0,4.5,6.,np.inf],labels = [1,2,3,4,5])
```

```python
df['incom_cat'].value_counts().sort_index().plot.bar(rot = 0)
plt.title('distribution of income categories')
plt.xlabel('Income Category')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['median_house_value'],axis = 1)
y = df['median_house_value']

X_train,X_test,y_train,y_test = train_test_split(X , y, test_size =0.2 , stratify = df['incom_cat'] , random_state= 42)

# Stratified split => keeps the same income distribution in train/test
```

```python
# we won’t use the income_cat column again
for data in (X_train,X_test):
  data.drop('incom_cat',axis =1 , inplace =True)
```

```python
train_data = X_train.join(y_train)
train_data
```

# Prepare the Data for Machine Learning Algorithms

 Feature Scaling and Transformation

```python
train_data['total_rooms'] = np.log1p(train_data['total_rooms'])
train_data['total_bedrooms'] = np.log1p(train_data['total_bedrooms'])
train_data['population'] = np.log1p(train_data['population'])
train_data['households'] = np.log1p(train_data['households'])
train_data.hist(bins =50 ,figsize=(15,8))
```

# Handling Text and Categorical Attributes

```python
train_data['ocean_proximity'].unique()
```

```python
train_data = pd.get_dummies( data = train_data ,columns=['ocean_proximity']).astype(int)
```

```python
train_data.head()
```

Create New Features

```python
train_data['bedrooms_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
```

```python
train_data['people_per_house'] = train_data['population'] / train_data['households']
```

```python
train_data['rooms_per_house'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize = (17,8))
sns.heatmap(data =train_data.corr() , annot = True , cmap = "YlGnBu")
plt.show()
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

x_train , y_train =  train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
test_data = X_test.join(y_test)
test_data['total_rooms'] = np.log1p(test_data['total_rooms'])
test_data['total_bedrooms'] = np.log1p(test_data['total_bedrooms'])
test_data['population'] = np.log1p(test_data['population'])
test_data['households'] = np.log1p(test_data['households'])

test_data = pd.get_dummies( data = test_data ,columns=['ocean_proximity']).astype(int)

test_data['bedrooms_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']

test_data['people_per_house'] = np.where(
    test_data['households'] == 0,
    0,
    test_data['population'] / test_data['households']
)

test_data['rooms_per_house'] = np.where(
    test_data['households'] == 0,
    0,
    test_data['total_rooms'] / test_data['households']
)
x_test , y_test =  test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
reg.score(x_test, y_test)
```
